```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
R = np.random.random([100,1]) +3
```

```python
R[0:5]
```

```python
plt.plot(R)
plt.show()
```

```python
np.mean(R)
```

```python
np.std(R)
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(R)
```

```python
scalar.mean_ # already computed mean and standard deviation
```

```python
Rscaled = scalar.transform(R)
```

```python
plt.plot(Rscaled) # before scale above plot and after scale plot same
```

```python
np.mean(Rscaled)
```
